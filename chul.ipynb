```python
import pandas as pd
```

```python
def custom_change(value):
  if value == '0%':
    return '결석'
  elif value == '50%':
    return '지각'
  elif value == '100%':
    return '출석'
```

```python
check4 = pd.read_csv('출결4.csv')
check4 = check4.dropna()
check4 = check4.drop(check4.columns[-1], axis=1)
# 4월 출결의 이름을 'O'을 '*'으로 바꾸기
check4['이름'] = check4['이름'].str.replace('O', '*')

# 출석률 바꾸기
for col in check4.columns[2:]:
    check4[col] = check4[col].apply(custom_change)

check4
```

```python
# 결석한 횟수를 계산
def count_absent(row):
    return (row == '결석').sum()
def count_late(row):
    return (row == '지각').sum()

check4['결석 횟수'] = check4.iloc[:, 2:].apply(count_absent, axis=1)
check4['지각 횟수'] = check4.iloc[:, 2:].apply(count_late, axis=1)

# 결과 출력
print(check4[['이름', '지각 횟수', '결석 횟수']])
```
